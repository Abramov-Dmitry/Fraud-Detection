```python
df = pd.read_csv("creditcard.csv")
df.head()
```

```python
# Проверка на отстутствующие значения
missing_data = df.isna().sum()
missing_data[missing_data > 0]
```

**Отсутствующих значений в данных нет**

```python
df.describe()
```

```python
# Переименование классов. 0: Не мошенничество, 1: Мошенничество
df.Class.replace({0: "Not Fraud", 1: "Fraud"}, inplace=True)
# Вывод количества классов
df["Class"].value_counts()
```

```python
df.loc[df["Amount"] == 0.0]
```

```python
print(f"Размерность данных перед удалением транзакций с суммой 0.0: {df.shape}")
df.drop(np.where(df["Amount"] == 0.0)[0], inplace=True)
df.drop_duplicates(inplace=True)
print(f"Размерность данных после удаления транзакций с суммой 0.0: {df.shape}")
```

```python
# Обучение случайного леса для того, чтобы найти количество признаков для 90%
# покрытия информации
rf = RandomForestClassifier(n_jobs=-1, random_state=42)
X = df.drop("Class", axis=1).values
y = df.Class.values
y = LabelEncoder().fit_transform(y)

rf.fit(X, y)
```

```python
# Визуализация вклада каждого признака
feature_importances = rf.feature_importances_
feature_importances = np.array(feature_importances)
pd.Series(feature_importances).sort_values().plot(kind="barh",
                                                  figsize=(16, 9),
                                                  fontsize=18,
                                                  color="green")
# График кумулятивной
plt.figure()
plt.subplots(figsize=(16, 6))
plt.xticks(ticks=list(range(len(feature_importances))))
plt.tick_params(axis='both', which='major', labelsize=18)
plt.axhline(0.9, color="red")
plt.axvline(20, color="red")
sns.lineplot(x=list(range(len(feature_importances))),
             y=np.cumsum(feature_importances),
             color="green",
             lw=3)
```

20 признаков покрывают >90% информации в выборке

```python
print(feature_importances * 100)
np.cumsum(sorted(feature_importances, reverse=True))
```

```python
# Сводка по мошенничествам, с суммой больше 0.00
amount_sum = df[(df.Class == "Fraud") & (df.Amount > 0.00)]["Amount"]\
	.value_counts()
amount_sum = amount_sum.reset_index().rename(columns={'index': 'Amount',
                                                      'Amount': 'Count_of_Amount'})

print(f"Средняя сумма мошенничества: {amount_sum.Amount.mean()}")
print(f"Самая частая сумма мошшеничества: {amount_sum.value_counts().max()}")
print(f"Максимальная сумма мошенничества: {amount_sum.Amount.max()}")
```

```python
# Изначальная размерность данных
df_without_outliers = df.copy()
# Удаление информации с выбросами, без учета времени, меток класса и суммы
for col in df.columns.drop(["Time", "Class", "Amount"]):
	# Межквартильный размах
	iqr = df[col].quantile(0.75) - df[col].quantile(0.25)

	lower_range = df[col].quantile(0.25) - (1.5 * iqr) # Нижняя граница
	upper_range = df[col].quantile(0.75) + (1.5 * iqr) # Верхняя граница
	# Удаление строк с обнаруженными выбросами
	df_without_outliers.drop(df_without_outliers
							 [(df_without_outliers[col] > upper_range) |
							  (df_without_outliers[col] < lower_range)].index,
							 inplace=True)

# Количество меток класса 0 и 1 в данных, с удаленными выбросами
count_labels = Counter(df_without_outliers["Class"])
# Процент оставшихся мошеннических случаев после удаления выбросов
class_1_without_outliers = count_labels['Fraud'] / df['Class']\
	[df['Class']=='Fraud'].count() * 100

print(f"Соотношение меток класса: {count_labels['Not Fraud']}/"
	  f"{count_labels['Fraud']}")
print(f"Осталось {class_1_without_outliers:0.2f}% меток класса 1")
```

**При удалении данных, превышающих 1.5IQR, остается слишком мало информации о мошенниках**
